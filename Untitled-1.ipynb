# Problema da Cobertura Mxima
## Descrio do problema

Imagine que uma prefeitura tem 4 opes de terrenos e queira construir 2 escolas novas, para atender uma populao de crianas localizadas em bairros em diferentes pontos da cidade. Consideramos que uma escola pode atender crianas que residem a uma distncia de at 3 km da escola.

Dadas as opes de terrenos (representados pelos quadrados no diagrama) e a distribuio da populao das crianas nos bairros (representados pelos crculos), quais locais vo garantir que a maior quantidade de crianas sejam atendidas?



## Modelagem do problema

#### Variveis:
* $x_{i} \in \{0, 1\}$ : Se o terreno i ser selecionado
* $y_{j} \in \{0, 1\}$ : Se o bairro j ser atendido por pelo menos uma escola

#### Funo Objetivo:
* Maximizar os bairros atendidos

#### Restries:
* No podemos abrir mais que duas escola
* Cada bairro s ser atendido se pelo menos um dos locais que o cobrem for selecionado

### A modelagem do exemplo acima seria:
$$\max y_{1} + y_{2} + y_{3} + y_{4} + y_{5} + y_{6} + y_{7} + y_{8} + y_{9} + y_{10}$$

$S.a.:$
$$x_{1} + x_{2} + x_{3} + x_{4} = 2$$
$$y_{1}  x_{3}$$
$$y_{2}  x_{2}$$
$$y_{3}  x_{1} + x_{4}$$
$$y_{4}  x_{4}$$
$$y_{5}  x_{2}$$
$$y_{6}  x_{4}$$
$$y_{7}  x_{2}$$
$$y_{8}  x_{2}$$
$$y_{9}  x_{3}$$
$$y_{10}  x_{1} + x_{3}$$
$$x_{1}, x_{2}, x_{3}, x_{4}, y_{1}, y_{2}, y_{3}, y_{4}, y_{5}, y_{6}, y_{7}, y_{8}, y_{9}, y_{10} \in \{0, 1\}$$

## Solucionando o problema usando Python MIP


```python
from mip import *

# Instanciando o modelo
model_MCP = Model(sense=MAXIMIZE, solver_name=CBC)
model_MCP.verbose = 0

# Adicionando as variveis dos locais a seres escolhidos e os bairros a serem cobertos
x = [model_MCP.add_var(var_type=BINARY, name = f"x_{i+1}") for i in range(4)]
y = [model_MCP.add_var(var_type=BINARY, name = f"y_{j+1}") for j in range(10)]

# Adicionando a funo objetivo
model_MCP.objective = xsum(y[i] for i in range(10))

#  Adicionando as restries
#  lembrando que, j que o ndice das listas comeam em 0, c1 vira c[0], l1 vira l[0], c2 vira c[1] e assim por diante
model_MCP += x[0]+x[1]+x[2]+x[3] == 2 
model_MCP += y[0] <= x[2]
model_MCP += y[1] <= x[1]
model_MCP += y[2] <= x[0]+x[3]
model_MCP += y[3] <= x[3]
model_MCP += y[4] <= x[1]
model_MCP += y[5] <= x[3]
model_MCP += y[6] <= x[1]
model_MCP += y[7] <= x[1]
model_MCP += y[8] <= x[2]
model_MCP += y[9] <= x[0]+x[2]

# Resolvendo o modelo
model_MCP.optimize()

print("Valor da soluo = ", model_MCP.objective_value)

print("\nLocais escolhidos:")
for v in model_MCP.vars: 
    if v.name[0] == "x" and v.x > 0.00001: 
        print(v.name)
print("\nBairros cobertos:")
for v in model_MCP.vars: 
    if v.name[0] == "y" and v.x > 0.00001: 
        print(v.name)
```

## Generalizando o problema

No Problema da Cobertura Mxima, precisamos escolher, entre um conjunto de locais candidatos, um subconjunto de tamanho determinado para estabelecer facilidades que atendam a maior quantidade de clientes, dado o raio de cobertura da facilidade em questo.

No exemplo dado, os locais candidatos eram os terrenos, as facilidades eram escolas e os clientes crianas. Porm, podemos tratar de diversos outros tipos de facilidades, como hospitais, delegacias, supermercados, antenas etc, e os clientes sero o grupo de interesse a quem se deseja atender.

Uma observao importante  que diferentes clientes podem ter diferentes demandas. No exemplo dado, consideramos que os bairros tm populao igual. Porm, podemos ter bairros mais populosos que outros. Para isso, podemos atribuir coeficientes diferentes (pesos) s variveis na funo objetivo, para levar isso em conta.

Assim, uma forma geral de representar o PCM pode ser:

#### Entrada:
* $L$: conjunto de locais candidatos
* $C$: conjunto de clientes
* $p$: nmero de facilidades a serem abertas
* $h_j$: demanda do cliente $j \in C$
* $d_{ij}$: distncia entre o local $i \in L$ e o cliente $j \in C$
* $R$: raio de cobertura das facilidades

#### Variveis:
* $x_{i} \in \{0, 1\}$ : Se o local $i$ ser selecionado
* $y_{j} \in \{0, 1\}$ : Se o cliente $j$ ser atendido por pelo menos uma facilidade

#### Funo Objetivo:
* Maximizar a soma das demandas dos clientes atendidos

#### Restries:
* Devemos selecionar apenas $p$ locais candidatos
* Cada cliente s ser atendido se pelo menos um dos locais que o cobrem for selecionado

#### Modelagem:
$$\max \sum_{j \in C} h_{j}*y_{j} $$

$S.a.:$
$$ \sum_{i \in L} x_{i} = p$$
$$ y_{j}  \sum_{i \in L: d_{ij} \leq R} x_{i}, \space \forall j \in C$$
$$x_{i} \in \{0, 1\}, \space \forall i \in L$$
$$y_{j} \in \{0, 1\}, \space \forall j \in C$$

```python
# Usando a generalizao, vamos criar uma funo que instancia um PCM

def create_MCP_model(n_loc: int, n_fac: int, n_cli: int, rad: float, demands: List[float], distances: List[List[float]]):
    """
    Parameters:
    -----------
    n_loc: int
        Nmero de locais candidatos
    n_fac: int
        Nmero de facilidades a serem construdas, ou o nmero de locais a serem escolhidos
    n_cli: int
        Nmero de clientes
    rad: float
        Raio de cobertura das facilidades
    demands: List[int]]
        Lista com a demanda de cada cliente
        demand[j-1] deve representar hj, ou seja, a demanda do cliente j
    distances: List[List[int]]
        Matriz de tamanho n_loc x n_cli, com as distncias entre cada cliente e cada local candidato
        distances[i-1][j-1] deve representar dij, ou seja, a distncia entre o local i e o cliente j

    Returns:
    --------
    model: mip.Model
        Modelo j instanciado e pronto para ser otimizado
    """

    # Instanciando o modelo
    model = Model(sense=MAXIMIZE, solver_name=CBC)
    model.verbose = 0

    # Adicionando as variveis dos locais a seres escolhidos e os clientes a serem cobertos
    x = [model.add_var(var_type=BINARY, name = f"x_{i+1}") for i in range(n_loc)]
    y = [model.add_var(var_type=BINARY, name = f"y_{j+1}") for j in range(n_cli)]

    # Adicionando a funo objetivo
        # que  o somatrio de hi*ci para todo i de 1 a n_cli
    model.objective = xsum(demands[j]*y[j] for j in range(n_cli))

    #  Adicionando as restries
        #  o somatrio de lj para todo j de 1 a n_loc deve ser igual a n_fac
    model += xsum(x[i] for i in range(n_loc)) == n_fac
        #  para todo i, ci deve ser menor ou igual ao somatrio de lj para todo j tal que dij <= rad
    for j in range(n_cli):
        model += y[j] <= xsum(x[i] for i in range(n_loc) if distances[i][j] <= rad)
    
    return model
```

```python
# Para o problema do exemplo:

locais = 4
facilidades = 2
clientes = 10
raio = 3

# Nesse caso, vamos considerar que a populao de crianas entre os bairros  constante
# Ou seja, a demanda dos clientes  a mesma
demandas = [1 for i in range(clientes)]

distancias = [[ 7.,   6.,   2.8,  6.,   8.7,  5.4,  8.,  10.2,  7.2,  2.5],
              [ 7.2,  2.9,  8.5,  4.8,  0.4,  9.,   2.2,  1.8,  9.,   7.5],
              [ 2.,   4.9,  6.,   6.5,  7.,   9.,   6.8,  8.5,  2.,   2.2],
              [ 8.5,  5.5,  2.8,  1.5,  6.,   1.5,  4.5,  6.6,  9.5,  5.]]

model = create_MCP_model(locais, facilidades, clientes, raio, demandas, distancias)
solution = model.optimize()

print("Status = ", solution)
print("Valor da soluo = ", model.objective_value)

print("\nLocais escolhidos:")
for v in model.vars: 
    if v.name[0] == "x" and v.x > 0.00001: 
        print(v.name)
print("\nClientes cobertos:")
for v in model.vars: 
    if v.name[0] == "y" and v.x > 0.00001: 
        print(v.name)
```
